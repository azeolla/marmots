```python
freq0 = np.std(np.real(spectrum[:,:,0]), axis=1)
freq1 = np.std(np.real(spectrum[:,:,1]), axis=1)
freq2 = np.std(np.real(spectrum[:,:,2]), axis=1)
freq3 = np.std(np.real(spectrum[:,:,3]), axis=1)
freq4 = np.std(np.real(spectrum[:,:,4]), axis=1)
```

```python
def random_rms(rms, N):
        """
        Produces a spectrum of noise.
        """
        
        num_freqs = rfftfreq(1024, 2*1e-9).size
        
        amplitude = np.random.rayleigh((rms*np.sqrt(num_freqs)), size=(N, num_freqs))
        phase = 2. * np.pi* np.random.random((N, num_freqs))
        noise = amplitude * np.cos(phase) + 1j * amplitude * np.sin(phase)
        
        time = irfft(noise, axis=1)

        return np.std(time, axis=1)
```
