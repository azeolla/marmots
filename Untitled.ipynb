```python
import poinsseta.geometry as geometry
import numpy as np
from poinsseta.constants import Re
import matplotlib.pyplot as plt
from shapely.geometry import Polygon, MultiPolygon
```

```python
theta_min = np.deg2rad(0)
theta_max = np.deg2rad(1)
phi_min = np.deg2rad(0)
phi_max = np.deg2rad(360)

true_area = geometry.spherical_segment_area(theta_min, theta_max, phi_min, phi_max, 0)
print(true_area)
```

```python
def project(latitude, longitude):
    earth_radius = Re
    lat_dist = np.pi * earth_radius / 180
    y = latitude * lat_dist 
    x = longitude * lat_dist * np.cos(np.radians(latitude))
    return x, y
```

```python
longitude = np.array([3, 3, -3, -3])
latitude = np.array([80, 78, 78, 80])
x,y = project(latitude, longitude)
polygon1 = Polygon(zip(x,y))
```

```python
longitude = np.array([5, 5, -1, -1])
latitude = np.array([81, 79, 79, 81])
x,y = project(latitude, longitude)
polygon2 = Polygon(zip(x,y))
```
